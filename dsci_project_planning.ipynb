## (3) **Exploratory Data Analysis and Visualization**

Each dataset will be tidied separately by renaming variables to lowercase and re-formatting timestamps.

```r
# Make all variables in the players data lowercase to ensure tidyness

# Rename columns

players_tidy <- players |>
  rename_with(tolower)

players_tidy
```

```r
# Re-format timestamps in the sessions data to ensure tidyness

sessions_tidy <- sessions |>
    mutate(start_time = dmy_hm(start_time),
           end_time = dmy_hm(end_time),
           original_start_time = as_datetime(original_start_time / 1000),
           original_end_time = as_datetime(original_end_time / 1000)
           )

sessions_tidy
```

#### Mean Values For All Quantitative Variables

```r
# For both the players and sessions datasets

means_players <- players |>
  summarise(
    mean_age = mean(Age, na.rm = TRUE),
    mean_played_hours = mean(played_hours, na.rm = TRUE),
  )

means_sessions <- sessions |>
    summarize(mean_origin_start = mean(original_start_time, na.rm = TRUE),
              mean_origin_end = mean(original_end_time, na.rm = TRUE)
              )

results_means <- bind_cols(means_players, means_sessions)
results_means
```

#### Planned Visualizations

##### Boxplot: Comparing played hours between subscribers and non-subscribers (to explore engagement patterns)

```r
options(repr.plot.width = 7, repr.plot.height = 7) 

hours_played_sub <- players |>
    ggplot(aes(x = subscribe, y = played_hours)) +
    geom_boxplot() +
    scale_x_discrete(labels = c("FALSE" = "Unsubscribed",
                                    "TRUE" = "Subscribed")) +
    labs(title = "Engagement Patterns by Subscription Status", 
         x = "Subscription Status", 
         y = "Hours Played") +
        theme(text = element_text(size = 15))

hours_played_sub
```

##### Bar Plot: Visualize experience level vs subscription rate

```r
options(repr.plot.width = 9, repr.plot.height = 9) 

experience_sub <- players |>
    ggplot(aes(x = experience, fill = subscribe)) +
    geom_bar(position = "fill") +
    scale_fill_manual(values = c("FALSE" = "red",
                                 "TRUE" = "blue"),
                      labels = c("FALSE" = "Unsubscribed",
                                     "TRUE" = "Subscribed")) +
    labs(title = "Subscription Status By Experience Level", 
         x = "Experience Level",
         y = "Percentage Of Players",
        fill = "Subscription Status") +
        theme(text = element_text(size = 15)) +
        theme(legend.position = "right", legend.direction = "vertical")
        
experience_sub
```

#### Insights Gained From Plots Above

The visualizations help show whether more experienced players or more active players are likely to subscribe, which reveals overall engagement patterns. In addition, the boxplot and proportion bars helps us identify extreme outliers such as players with extremely high playtime, which may impact averages or indicate a small group of highly active users.
